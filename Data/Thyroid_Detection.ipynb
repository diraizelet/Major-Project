### Logistic regression

The goal of logistic regression is to find the best weights (w_1, w_2) that will make the predicted probability close to 1 for customers who buy the product and close to 0 for customers who don’t.

The model uses a technique called maximum likelihood estimation (MLE) to find the weights that maximize the likelihood of observing the given outcomes (whether a customer bought the product or not) based on their age and income.

```python
def log_classifier(X_train,X_test,y_train,y_test):
    log_model=LogisticRegression()
    log_model.fit(X_train,y_train)
    log_pred=log_model.predict(X_test)
    cm=confusion_matrix(y_test,log_pred)
    acc = accuracy_score(log_pred,y_test)
    return (f'Accuracy_Score: {acc}\n Train Score: {log_model.score(X_train,y_train)}\n Test Score: {log_model.score(X_test,y_test)}\n Confusion Matrix:\n {cm}')
```

### SVM

In SVM, the goal is to find a hyperplane that divides the data points of different classes. A hyperplane is a decision boundary that separates the feature space into two halves.

```python
def svm_classifier(X_train,X_test,y_train,y_test):
    classifier_svm=SVC(kernel='rbf',random_state=0)
    classifier_svm.fit(X_train,y_train)
    svm_pred=classifier_svm.predict(X_test)
    cm=confusion_matrix(y_test,svm_pred)
    acc = accuracy_score(svm_pred,y_test)
    return (f'Accuracy_Score: {acc}\n Train Score: {classifier_svm.score(X_train,y_train)}\n Test Score: {classifier_svm.score(X_test,y_test)}\n Confusion Matrix:\n {cm}')
```

### knn

we define how many nodes we use for the training and then perform training based on how the nodes results are and how the data is structured.
it doesnot work for this as there is not really much correaltion of each and every node.

```python
def knn_classifier(X_train,X_test,y_train,y_test):
    classifier_knn=KNeighborsClassifier(metric='minkowski',p=2)
    classifier_knn.fit(X_train,y_train)
    knn_pred=classifier_knn.predict(X_test)
    cm=confusion_matrix(y_test,knn_pred)
    acc = accuracy_score(knn_pred,y_test)
    return (f'Accuracy_Score: {acc}\n Train Score: {classifier_knn.score(X_train,y_train)}\n Test Score: {classifier_knn.score(X_test,y_test)}\n Confusion Matrix:\n {cm}')
```

### Decision Tree

it is a tree where each and everylevel is a question about the attribute and makes a decision based on the value.

```python
def tree_classifier(X_train,X_test,y_train,y_test):
    classifier_tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
    classifier_tree.fit(X_train,y_train)
    dt_pred=classifier_tree.predict(X_test)
    cm=confusion_matrix(y_test,dt_pred)
    acc = accuracy_score(dt_pred,y_test)
    return (f'Accuracy_Score: {acc}\n Train Score: {classifier_tree.score(X_train,y_train)}\n Test Score: {classifier_tree.score(X_test,y_test)}\n Confusion Matrix:\n {cm}')
```

### Random Forest

each and every attribute has its own tree and the tree makes a decision depending on the value of the attribute and the collective decisions are stored and then the final decision is a majority label.

```python
def forest_classifier(X_train,X_test,y_train,y_test):

    # create an instance of a Random Forest classifier
    classifier_forest=RandomForestClassifier(criterion='entropy',random_state=0)
    # training the Random Forest model on the sampled train dataset
    classifier_forest.fit(X_train,y_train)
    # pedicting the labels on the test set
    rf_pred=classifier_forest.predict(X_test)
    cm=confusion_matrix(y_test,rf_pred)
    acc = accuracy_score(rf_pred,y_test)
    #cl_report = classification_report(y_test,rf_pred)
    return (f'Accuracy_Score: {acc}\n Train Score: {classifier_forest.score(X_train,y_train)}\n Test Score: {classifier_forest.score(X_test,y_test)}\n Confusion Matrix:\n {cm}')
```

```python
def print_score(X_train, X_test, y_train, y_test):
    print("Logitsic Regression:\n")
    result0=log_classifier(X_train, X_test, y_train, y_test)
    print(result0)
    print("-"*100)
    print()

    print("SVM:\n")
    result1=svm_classifier(X_train, X_test, y_train, y_test)
    print(result1)
    print("-"*100)
    print()

    print("KNN:\n")
    result2=knn_classifier(X_train, X_test, y_train, y_test)
    print(result2)

    print("-"*100)
    print()

    print("Decision Tree:\n")
    result4=tree_classifier(X_train, X_test, y_train, y_test)
    print(result4)

    print("-"*100)
    print()

    print("Random Forest:\n")
    result5=forest_classifier(X_train, X_test, y_train, y_test)
    print(result5)

print_score(X_train, X_test, y_train, y_test)
```

*From the above models,it is clear that Random forest is the best models for this dataset.*

```python
import pickle
import numpy as np

def load_model(file_path):
    """Loads the thyroid detection model from the given file path."""
    try:
        with open(file_path, 'rb') as file:
            model = pickle.load(file)
        print("Model loaded successfully.")
        return model
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except Exception as e:
        print(f"An error occurred while loading the model: {e}")
        return None

def predict_thyroid_status(model, test_data):
    """
    Predicts thyroid status based on the test data.
    Parameters:
        model: Loaded thyroid detection model.
        test_data: A list or numpy array containing test data (features).
    Returns:
        Predicted result from the model.
    """
    try:
        # Convert the test data to a NumPy array and reshape for a single prediction
        test_data = np.array(test_data).reshape(1, -1)
        prediction = model.predict(test_data)
        return prediction[0]
    except Exception as e:
        print(f"An error occurred during prediction: {e}")
        return None

# Path to your trained model file
model_path = "/content/thyroid_model.pkl"

# Load the model
thyroid_model = load_model(model_path)

if thyroid_model:
    # Constructing the test data with 24 features
    # Features in order: age, sex, on_thyroxine, query_on_thyroxine, on_antithyroid_medication,
    # sick, pregnant, thyroid_surgery, I131_treatment, query_hypothyroid, query_hyperthyroid,
    # lithium, goitre, tumor, hypopituitary, psych, T3, TT4, T4U, FTI,
    # referral_source_SVHC, referral_source_SVHD, referral_source_SVI, referral_source_other

    test_data = [
    50,    # age: 50 years old (middle-aged adult)
    0,     # sex: female (0 for female)
    0,     # on_thyroxine: not on thyroid medication
    0,     # query_on_thyroxine: no history of questioning thyroid medication
    0,     # on_antithyroid_medication: not on anti-thyroid medication
    0,     # sick: not sick
    0,     # pregnant: not pregnant
    0,     # thyroid_surgery: no thyroid surgery
    0,     # I131_treatment: no I131 treatment
    1,     # query_hypothyroid: suspicion of hypothyroidism
    0,     # query_hyperthyroid: no suspicion of hyperthyroidism
    0,     # lithium: not on lithium
    0,     # goitre: no goitre (secondary hypothyroidism does not usually cause goitre)
    0,     # tumor: no thyroid tumor
    1,     # hypopituitary: confirmed pituitary dysfunction
    0,     # psych: no psychiatric issues
    0.5,   # T3: significantly below normal (severe deficiency, ~0.5–0.8 ng/mL)
    30,    # TT4: very low (~30 µg/dL; normal is 5–12 µg/dL)
    0.7,   # T4U: slightly low (often 0.7–0.8 in secondary hypothyroidism)
    35,    # FTI: very low (~35, normal is 60–120)
    0,     # referral_source_SVHC: no referral from SVHC
    0,     # referral_source_SVHD: no referral from SVHD
    1,     # referral_source_SVI: referral from a specialist (SVI)
    0      # referral_source_other: no referral from other sources
]




    # Get prediction
    result = predict_thyroid_status(thyroid_model, test_data)

    if result is not None:
        print(f"Predicted Thyroid Class: {result}")
        if result == 0:  # Assuming '0' indicates "no thyroid issue"
            print("The model predicts no thyroid issue.")
        else:
            print("The model predicts a thyroid issue.")
    else:
        print("Prediction failed.")
```

```python
import pickle

def load_model(file_path):
    """Loads the thyroid detection model from the given file path."""
    try:
        with open(file_path, 'rb') as file:
            model = pickle.load(file)
        print("Model loaded successfully.")
        return model
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except Exception as e:
        print(f"An error occurred while loading the model: {e}")
        return None

def get_required_features(model):
    """
    Retrieves the feature names expected by the model.
    Returns:
        A list of feature names if available; otherwise, None.
    """
    try:
        # Check if the model has feature names (common for scikit-learn pipelines or models)
        if hasattr(model, "feature_names_in_"):
            return model.feature_names_in_.tolist()
        elif hasattr(model, "feature_names"):  # Custom attribute in some cases
            return model.feature_names
        else:
            print("Feature names could not be extracted. Ensure the model has labeled feature metadata.")
            return None
    except Exception as e:
        print(f"An error occurred while retrieving feature names: {e}")
        return None

# Path to your trained model file
model_path = "/content/thyroid_model.pkl"

# Load the model
thyroid_model = load_model(model_path)

if thyroid_model:
    # Get required features
    required_features = get_required_features(thyroid_model)

    if required_features:
        print("The model expects the following features (in order):")
        for idx, feature in enumerate(required_features, 1):
            print(f"{idx}. {feature}")

        # Optionally, create a template for the test data
        print("\nTemplate for test data (fill values in order):")
        print([f"{feature}: <value>" for feature in required_features])
    else:
        print("Unable to determine the required features for this model.")
else:
    print("Model loading failed.")
```
